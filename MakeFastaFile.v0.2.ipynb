DNA塩基配列が書かれたファイルを読み込んで、Fasta形式に変換して保存するスクリプト

```python
# マジックコマンドでセル全体の実行時間を計測 # Unix系OSの場合のみ
%time

# ファイル読み込み
infile = "Test1.tsv"

# DNA塩基配列の通し番号
i = 1

with open (infile, 'r') as inf:
    # 1行読み込み
    record = inf.readline()
    # Fasta形式に変換
    record = ">Seq" + str(i) + "\n" + record
    # 確認
    print (record)
```

```python
# マジックコマンドでセル全体の実行時間を計測 # Unix系OSの場合のみ
%time
# ファイルを1行ずつ読み込む場合の処理時間

# ファイル読み込み
infile = "Test1.tsv"

# DNA塩基配列の通し番号
i = 1

with open (infile, 'r') as inf:
    
    # ファイル全体を1行ずつ読み込む
    for record in inf:
        # 1行読み込み
        record = inf.readline()
        # Fasta形式に変換
        record = ">Seq" + str(i) + "\n" + record
        # 確認
        print (record)
        i+= 1

```

```python
# マジックコマンドでセル全体の実行時間を計測 # Unix系OSの場合のみ
%time
# ファイルを1行ずつ読み込んで、1行ずつ書き込む処理時間


# ファイル読み込み
infile = "Test1.tsv"

# ファイル書き込み
outfile = "FASTA_" + infile

# DNA塩基配列の通し番号
i = 1

with open (infile, 'r') as inf:
    
    with open (outfile, 'w') as out: # 読み込みファイルから読み出しながら、処理結果を書き込みファイルに書き出せる
    
        # ファイル全体を1行ずつ読み込む
        for record in inf:
            # 1行読み込み
            record = inf.readline()
            # Fasta形式に変換
            record = ">Seq" + str(i) + "\n" + record
            # 確認
            #print (record)
        
            # 書き込み
            out.write(record)
        
            # 番号をインクリメント
            i+= 1
        
```

```python
# マジックコマンドでセル全体の実行時間を計測 # Unix系OSの場合のみ
%time
# ファイル全体をまとめて読み込んで、1行ずつ書き出す処理時間


# ファイル読み込み
infile = "Test1.tsv"

with open (infile, 'r') as inf:
    
    # ファイルの内容を行ごとのリストとして読み込む
    records = inf.readlines()

# DNA塩基配列の通し番号
i = 1

# 確認
for record in records:
    # FASTA形式
    print(">Seq", str(i), "\n", sep='')
    print(record)
    i += 1
```

```python
# マジックコマンドでセル全体の実行時間を計測 # Unix系OSの場合のみ
%time
# ファイル全体をまとめて読み込んで、1行ずつ書き込む処理時間

# ファイル読み込み
infile = "Test1.tsv"
# ファイル書き込み
outfile = "FASTA_" + infile

with open (infile, 'r') as inf:
    
    # ファイルの内容を行ごとのリストとして読み込む
    records = inf.readlines()


# DNA塩基配列の通し番号
i = 1

# リストをファイルに書き込む
with open (outfile, 'w') as out:
    
    for record in records:
        # FASTA形式
        fasta_id = ">Seq" + str(i) + "\n"
        out.write(fasta_id)
        out.write(record)

        i += 1
```
